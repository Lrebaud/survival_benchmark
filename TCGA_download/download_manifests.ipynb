```python
from glob import glob
import os
from tqdm import tqdm
```

```python
for f in tqdm(glob('/mnt/data/lrebaud/Documents/ICARE_TCGA/download/*')):
    
    if os.path.isfile(f+'/tpm.csv'):
        continue
    print(f)
    path_rna_seq = f+'/'+'RNA-seq'
    try:
        os.mkdir(path_rna_seq)
    except:
        pass
    manifest_file = glob(f+'/*.txt')[0]
    command = 'cd '+path_rna_seq + ' && /mnt/data/lrebaud/Documents/ICARE_TCGA/gdc-client download -m '+manifest_file
    os.system(command)
```
