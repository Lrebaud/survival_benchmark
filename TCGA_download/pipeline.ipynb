```python
import pandas as pd
import numpy as np
```

```python
clinical = pd.read_excel('clinical/clinical_TCGA.xlsx', index_col='bcr_patient_barcode')
tpm = pd.read_csv('demo/merged_RNAseq/tpm.csv')
gene_annotations = pd.read_csv('demo/merged_RNAseq/gene_annotations.csv')
```

```python
gene_annotations = gene_annotations[gene_annotations['gene_type']=='protein_coding']
```

```python
tpm = tpm.T
tpm.columns = tpm.iloc[0]
tpm = tpm.iloc[1:]
tpm = tpm[tpm['Sample Type']=='Primary Tumor']
tpm.set_index('Case ID', inplace=True)
tpm = tpm.drop(columns=['Sample Type'])
assert len(tpm.index.unique()) == len(tpm)

tpm = tpm[gene_annotations['gene_id']]
# tpm.columns = gene_annotations['gene_name']
```

```python
common_patient = list(set(tpm.index)&set(clinical.dropna(how='any', subset=['OS', 'OS.time']).index))
tpm = tpm.loc[common_patient]
clinical = clinical.loc[common_patient, ['OS', 'OS.time']]
```

```python
tpm = tpm.astype('float64')
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class OmicsPreprocessor(TransformerMixin):

    def __init__(self, counts_threshold): #, gene_names):
        self.counts_threshold = counts_threshold
#         self.gene_names = gene_names
#         self.tuple_genes = pd.MultiIndex.from_frame(self.gene_names)
        self.selected_genes_ = None

    def fit(self, X, y=None):
        # 1. Transform data into pandas dataframe for preprocessing operations
#         temp = pd.DataFrame(X.copy().T, index=self.tuple_genes)
        temp = X.copy().T

        # 2. Filter low-expressed genes
        temp_filtered = temp.loc[(temp > 0).sum(axis=1) > self.counts_threshold*temp.shape[1]]
        
        # 3. Drop duplicates based on lowest standard deviation
#         def aggregate_dup(data):
#             data['std'] = np.log(data.copy() + 1).std(axis=1)
#             data = (data.sort_values(by='std', ascending=False)
#                     .groupby(axis=0, level=0)
#                     .first()
#                     .drop(columns='std')
#                     )
            
#             return data

#         temp_filtered = aggregate_dup(temp_filtered)
        self.selected_genes_ = temp_filtered.index.values
        return self

    def transform(self, X):
        # 1. Transform data into pandas dataframe
        return np.log(X[self.selected_genes_] + 1)
    
tr = OmicsPreprocessor(counts_threshold=0.2)
tr.fit(tpm)
tr_tpm = tr.transform(tpm)
```

```python
from sksurv.util import Surv
y = Surv.from_arrays(event=np.array(clinical['OS']).astype('bool'), time=clinical['OS.time'])
```

```python
from icare.survival import IcareSurvival
from icare.metrics import harrell_cindex_scorer
```

```python
ic = IcareSurvival()
ic.fit(X=tr_tpm.iloc[:, :300], y=y)

harrell_cindex_scorer(ic, tr_tpm.iloc[:, :300], y)
```
